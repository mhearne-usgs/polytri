```python
import matplotlib.pyplot as plt
from polytri import polytri
```

```python
xpoly = [0,10,12,20,13,10,12,14,8,6,10,7,0,1,3,5,-2,5]
ypoly = [0,7,3,8,17,12,14,9,10,14,15,18,16,13,15,8,9,5]
triangles = polytri.getTriangles(xpoly,ypoly)
xpoly.append(xpoly[0])
ypoly.append(ypoly[0])
print '%i triangles found:' % len(triangles)
for triangle in triangles:
    print triangle
```

```python
fig1 = plt.figure()
plt.plot(xpoly,ypoly,'b');
plt.title('Integer Polygon');
fig2 = plt.figure()
plt.hold(True)
for triangle in triangles:
    xv,yv = triangle
    xv = list(xv)
    yv = list(yv)
    xv.append(xv[0])
    yv.append(yv[0])
    plt.plot(xv,yv,'rx');
    plt.plot(xv,yv,'r');
plt.title('Integer Vertices');
```
